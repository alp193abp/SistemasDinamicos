### Introdução

O cálculo fracionário é uma ferramenta que é bastante usada na modelagem matemática, em grande parte na modelagem de fenômenos físicos. Além disso, muitas áreas possuem diversos fenômenos que são descritos por dinâmicas com corportamentos discretos. Assim, pesquisadores frequentemente tentam desenvolver métodos e teorias do cálculo fracionário para casos discretos.

O cálculo fracinário discreto (em inglês, DFC - discrete fractional calculus) busca descrever dinâmicas em tempo discreto.

### Preliminares

**Definição 2.1**:Seja $f(t)$ uma função de classe $C$, isto é, continua por partes em $(t_0, + \infty)$ e integrável em qualquer subntervalo finito de $(t_0, + \infty)$. Então para $t>0$, a integral de Riemann-Liouville de $f(t)$ de ordem $\alpha$ é definida como

$_{t_0}I_t^{\alpha}f(t) = \frac{1}{\Gamma(\alpha)}\int_{t_0}^t (t-\tau)^{\alpha-1} f(\tau) d\tau$

onde $\alpha$ é um número real positivo e $\Gamma( \cdot )$ é a função gama.


**Definição 2.2** Seja $\alpha$ um número real positivo, $m-1 < \alpha \leq m$, $m \in \mathbb{N}^+$, e existe $f^{(m)}(t)$, além disso ela é de classe $C$. Então existe uma derivada fracionária de Caputo de $f(t)$ de ordem $\alpha$ e ela é definida como

$ _0 ^C D_t^{\alpha} f(t) = _0I_t^{m-\alpha} f^{(m)}(t), \hspace{0.5cm}t > 0 .$


Para a derivada de Caputo da função $t^{\mu}$, $\mu>0$, se $0\leq m-1 < \alpha \leq m <\mu+1$, então nós temos

$ _0 ^C D_t^{\alpha} t^{\mu} = \frac{\Gamma(\mu+1)}{\Gamma(\mu-\alpha+1)}t^{\mu-\alpha}, \hspace{0.4cm} t>0 .$

Considerando o DFC, a definição da função $f(t)$ é mudada como uma sequência $f(n)$. Seja $\mathbb{N}_a = \{a, a+1, a+2,\dots\}$ a escala de tempo isolada e $a \in \mathbb{R}$ fixo. O operador de diferença $\Delta$ é definido como $\Delta f(n) = f(n+1) - f(n)$.

**Definição 2.3** Seja $\mu: \mathbb{N}_a \rightarrow \mathbb{R}$ e $ v > 0$ seja dado. Então a soma fracionária de ordem $v$ é definida por

$\Delta_a^{-v} u(t) := \frac{1}{\Gamma(v)}\sum_{s=a}^{t-v} (t-\sigma(s))^{(v-1)}u(s), \hspace{0.4cm} t \in \mathbb{N}_{a+v}$

onde $a$ é o ponto inicial, $\sigma(s) = s + 1$ e $t^{(v)}$ é a função falling definida por

$t^{(v)} = \frac{\Gamma(t + 1)}{\Gamma(t + 1 - v)} .$

**Definição 2.4** Para $v > 0$, $v \notin \mathbb{N}$ e $u(t)$ definida em $\mathbb{N}_a$, a diferença delta tipo Caputo é definida por

$^C \Delta_a^v u(t) := \Delta_a^{-(m-v)} \Delta^m u(t) = \frac{1}{\Gamma(m-v)}\sum_{s=a}^{t-(m-v)} (t-\sigma(s))^{(m-v-1)}u(s), \hspace{0.4cm} t \in \mathbb{N}_{a+m-v}, m = [v] + 1,$

onde $v$ é a ordem de diferença.


**Definição 2.5** Para a equação de diferença fracionária delta 

$^C \Delta_a^v u(t) = f(t+v-1, u(t+v-1))$, $\Delta^k u(a) = u_k$, $m = [v] + 1$, $k = 0, \dots, m-1$

a equação integral discreta equivalente pode ser obtida por

$u(t) = u_0(t) + \frac{1}{\Gamma(v)}\sum_{s=a+m-v}^{t-v} (t-\sigma(s))^{(v-1)} \dot f(s+v-1,u(s+v-1)), \hspace{0.4cm} t \in \mathbb{N}_{a+m}$

onde a iteração inicial é 

$u_0(t) = \sum_{k=0}^{m-1}\frac{(t-a)^{(k)}}{k!}\Delta^k u(a) .$

### Mapas fracionários do seno e mapa padrão

Mapa do seno unidimensional $x_{n+1} = x_n + \mu \sin(x_n)$ onde $\mu$ é a amplitude dos pulsos no movimento da equação.

Reescrevendo a equação $\Delta x(n) = \mu \sin(x(n))$.

Do cálculo fracionário discreto, o fracionário pode ser dado por $^C \Delta_a^v x(t) = \mu \sin(x(t+v-1)), \hspace{0.3cm}  0 < v < 1, \hspace{0.3cm} t \in \mathbb{N}_{a+1-v},$ onde $v$ é a ordem de diferença.

O mapa bidimensional padrão (mapa de Chirikov–Taylor ou mapa padrão de Chirikov) é expresso por

$ x_{n+1} = x_n - K \sin(y_n) \\
  y_{n+1} = y_n + x_{n+1}
$

Este mapa descreve a superfície de Poincaré da seção do movimento de um sistema mecânico simples, conhecido como rotor (ou rotador) retrocedido (em inglês, kicked rotator) [2,3] . Temos que $x(n)$ e $y(n)$ são, respectivamente, o momento angular e a coordenada (posição angular). Eles são tomado módulo $2\pi$. E o $K$ mede a intensidade da força de retrocesso [2].

Considerando a generalização fracionária do momento $x(n)$, temos que, o mapa padrão pode ser expresso como um fracional

$ ^C \Delta_a^v x(t) = - K \sin(y(t+v-1)), \hspace{0.3cm} 0 < v \leq 1, \hspace{0.3cm} t \in \mathbb{N}_{a+1-v},\\
y(n) = y(n-1) + x(n)
$

### Caos em mapas fracionários discretos

Do **Teorema 2.5**, tem-se que a forma integral discreta para $0 < v < 1$, pode ser escrita como

$u(t) = u(a) + \sum_{s=a+1-v}^{t-v} (t-\sigma(s))^{(v-1)} * f(s+v-1,u(s+v-1)), \hspace{0.4cm} t \in \mathbb{N}_{a+1},$

onde $\frac{(t-\sigma(s))^{(v-1)}}{\Gamma(s)}$ é a função de kernel discreto e $(t-\sigma(s))^{(v-1)}=\frac{\Gamma(t-s)}{\Gamma(t-s+1-v)}$. Então, podemos expressar a formula numérica explicitamente:

$u(n) = u(a) + \frac{1}{\Gamma(v)} \sum_{j=1}^n \frac{\Gamma(n-j+v)}{\Gamma(n-j+1)} f(j-1,u(j-1)). \\ $


Assim, temos que a forma numérica explicita para o seno é dada por:

$x(n) = x(a) + \frac{\mu}{\Gamma(v)} \sum_{j=1}^n \frac{\Gamma(n-j+v)}{\Gamma(n-j+1)} \sin(x(j-1)) .$

E para o mapa padrão fracionário:

$x(n) = x(a) -\frac{K}{\Gamma(v)} \sum_{j=1}^n \frac{\Gamma(n-j+v)}{\Gamma(n-j+1)} \sin(y(j-1)) \\
y(n) = y(n-1) + x(n), \hspace{0.3cm} n\geq 1,$

onde $x(n)$ e $y(n)$ são tomados módulo $2\pi$. O momento $x(n)$ depende das informações passada $x(0), \dots, x(n-1)$ e isto é chamado de efeito da memória discreta do sistema.

### Simulações e resultados

```python
#Importando pacotes
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
v = 1
a = 0
x_0 = 0.3
n = 165
μ = 1

data_mu = np.linspace(1,4.5,450)
data_x150 = np.zeros([len(data_mu),40])

for k in range(0,len(data_mu)):
    data = np.zeros(n+1)
    Γ_v = math.gamma(v)
    data[0] = x_0

    for i in range(1,len(data)):
        aux = 0
        for j in range(1,i+1):
            aux += math.gamma(i-j+v)/math.gamma(i-j+1)*math.sin(data[j-1]) 
        data[i] = (x_0 + data_mu[k]/Γ_v * aux)%(2*math.pi)


    for i in range(0,40):
        data_x150[k,i] = data[126+i]
```

```python
fig = plt.figure()
plt.plot(data_mu, data_x150, '.', color = 'black')
plt.xlabel('μ')
plt.ylabel('x(n)')
plt.grid()
fig.set_figheight(15)
fig.set_figwidth(15)
plt.show()
```

```python
v = 0.8
a = 0
x_0 = 0.3
n = 165
μ = 1

data_mu = np.linspace(1,4.5,450)
data_x150 = np.zeros([len(data_mu),40])

for k in range(0,len(data_mu)):
    data = np.zeros(n+1)
    Γ_v = math.gamma(v)
    data[0] = x_0

    for i in range(1,len(data)):
        aux = 0
        for j in range(1,i+1):
            aux += math.gamma(i-j+v)/math.gamma(i-j+1)*math.sin(data[j-1]) 
        data[i] = (x_0 + data_mu[k]/Γ_v * aux)%(2*math.pi)
    for i in range(0,40):
        data_x150[k,i] = data[126+i]
```

```python
fig = plt.figure()
plt.plot(data_mu, data_x150, '.', color = 'black')
plt.xlabel('μ')
plt.ylabel('x(n)')
plt.grid()
fig.set_figheight(15)
fig.set_figwidth(15)
plt.show()
```

```python
v = 1
a = 0
x_0 = 3.5
y_0 = 1.8
n = 150
K = 2

data_x = np.zeros(n+1)
data_y = np.zeros(n+1)
KΓ_v = K*math.gamma(v)

data_x[0] = x_0 
data_y[0] = y_0

for i in range(1,len(data_x)):
    aux = 0
    for j in range(1,i+1):
        aux += math.gamma(i-j+v)/math.gamma(i-j+1)*math.sin(data_y[j-1]) 
    data_x[i] = (x_0 - KΓ_v * aux)%(2*math.pi)
    data_y[i] = (data_y[i-1] + data_x[i])%(2*math.pi)
```

```python
fig = plt.figure()
plt.plot(np.linspace(0,151,151), data_x)
plt.plot(np.linspace(0,151,151), data_x, '*')
plt.xlabel('n')
plt.ylabel('x(n)')
fig.set_figheight(8)
fig.set_figwidth(15)
plt.grid()
plt.show()
```

```python
fig = plt.figure()
plt.plot(np.linspace(0,151,151), data_y)
plt.plot(np.linspace(0,151,151), data_y, '*')
plt.xlabel('n')
plt.ylabel('y(n)')
fig.set_figheight(8)
fig.set_figwidth(15)
plt.grid()
plt.show()
```

```python
v = 1
a = 0
n = 165
K = 1.5
geracoes = 50

data_x = np.zeros([n+1,geracoes])
data_y = np.zeros([n+1,geracoes])
KΓ_v = K*math.gamma(v)

for k in range(0,geracoes):
    x_0 = 2*math.pi*np.random.rand()
    y_0 = 2*math.pi*np.random.rand()
    data_x[0,k] = x_0
    data_y[0,k] = y_0
    for i in range(1,len(data_x)):
        aux = 0
        for j in range(1,i+1):
            aux += math.gamma(i-j+v)/math.gamma(i-j+1)*math.sin(data_y[j-1,k]) 
        data_x[i,k] = (x_0 - KΓ_v * aux)%(2*math.pi)
        data_y[i,k] = (data_y[i-1,k] + data_x[i,k])%(2*math.pi)
```

```python
fig = plt.figure()
plt.plot(data_x, data_y,'.')
plt.xlabel('x(n)')
plt.ylabel('y(n)')
fig.set_figheight(15)
fig.set_figwidth(15)
plt.grid()
plt.show()
```
